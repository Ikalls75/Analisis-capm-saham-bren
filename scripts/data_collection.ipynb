```python
import yfinance as yf
```

```python
tickers = "^JKSE BREN.JK"

data = yf.download(tickers, start="2023-10-09", end="2025-10-05", interval='1d', multi_level_index=False)

data
```

```python
Clean_data = data['Close'].rename(columns={'^JKSE': 'Close_IHSG', 'BREN.JK': 'Close_BREN'})

Clean_data = Clean_data[['Close_IHSG', 'Close_BREN']]

Clean_data.index.name = 'Date'
Clean_data.columns.name = None

Clean_data
```

```python
print("\nMising values:\n", data.isnull().sum().values.sum())
```

```python
Clean_data.to_excel('Historical_Data.xlsx', index=True)
```
